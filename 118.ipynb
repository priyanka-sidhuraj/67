```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("/content/118.csv")


fig = px.scatter(df, x="petal_size", y="sepal_size")
fig.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt #to plot the elbow line
import seaborn as sns # to plot the clusters and centroids(k)

X = df.iloc[:, [0, 1]].values  #acessing the values of the dataset

wcss=[]

for i in range(1,10):
   kmeans = KMeans(i)
   kmeans.fit(X)  #it accepts an input for the sample data ( X )
   wcss_iter = kmeans.inertia_ # eculideian distance ,calculating the distance between the data points and the mean value
   wcss.append(wcss_iter)

number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(X)


sns.scatterplot(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color = 'yellow', label = 'Cluster 1')
sns.scatterplot(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color = 'blue', label = 'Cluster 2')
sns.scatterplot(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color = 'green', label = 'Cluster 3')

sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids')

plt.title('Clusters of Flowers')
plt.xlabel('Petal Size')
plt.ylabel('Sepal Size')
plt.show()
```
